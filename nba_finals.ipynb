```python
import pandas as pd

file_path = '/content/championsdata.csv'

try:
    df_champions = pd.read_csv(file_path)
    print(f"Данные успешно загружены из: {file_path}\n")
except FileNotFoundError:
    print(f"Ошибка: Файл '{file_path}' не найден. Пожалуйста, убедитесь, что файл находится в папке '/content/'.\n")
    df_champions = None # Устанавливаем df_champions в None, чтобы избежать ошибок далее

if df_champions is not None:
    print("\nПервые 5 строк данных из championsdata.csv:")
    display(df_champions.head())

    print("\nПоследние 5 строк данных из championsdata.csv:")
    display(df_champions.tail())

    df_champions.info()

    print(df_champions.dtypes)
```

```python
print("\nОписательная статистика для df_champions:\n")
display(df_champions.describe())
```

```python
import pandas as pd

if df_champions is not None:
    print("--- 1. Среднее и медиана по группам (например, по году для очков, передач, подборов) ---")
    grouped_by_year_avg = df_champions.groupby('Year')[['PTS', 'AST', 'TRB']].mean().reset_index()
    print("\nСредние значения очков, передач и подборов по годам:")
    display(grouped_by_year_avg.head())

    grouped_by_year_median = df_champions.groupby('Year')[['PTS', 'AST', 'TRB']].median().reset_index()
    print("\nМедианные значения очков, передач и подборов по годам:")
    display(grouped_by_year_median.head())

    print("\n--- 2. Подсчёт количества (например, количество игр каждой команды) ---")
    team_game_counts = df_champions.groupby('Team')['Game'].count().reset_index()
    team_game_counts.rename(columns={'Game': 'Total_Games'}, inplace=True)
    print("\nКоличество игр для каждой команды:")
    display(team_game_counts.sort_values(by='Total_Games', ascending=False).head())

    print("\n--- 3. Суммирование (например, суммарные очки и броски каждой команды) ---")
    team_total_stats = df_champions.groupby('Team')[['PTS', 'FG', 'FGA']].sum().reset_index()
    print("\nСуммарные очки, забитые и брошенные броски для каждой команды:")
    display(team_total_stats.sort_values(by='PTS', ascending=False).head())

    print("\n--- 4. Построение сводной таблицы (например, среднее количество очков по годам и командам) ---")
    pivot_table_pts = df_champions.pivot_table(index='Year', columns='Team', values='PTS', aggfunc='mean')
    print("\nСводная таблица средних очков по годам и командам:")
    display(pivot_table_pts.head())

else:
    print("DataFrame 'df_champions' не найден. Пожалуйста, загрузите данные сначала.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

if df_champions is not None:
    sns.set_style("whitegrid")

    print("\n--- 1. Гистограммы распределений --- ")
    numerical_cols = ['PTS', 'FG', 'AST', 'TRB', 'TOV']
    plt.figure(figsize=(15, 10))
    for i, col in enumerate(numerical_cols, 1):
        plt.subplot(2, 3, i) # 2 строки, 3 столбца
        sns.histplot(df_champions[col], kde=True)
        plt.title(f'Распределение {col}')
        plt.xlabel(col)
        plt.ylabel('Частота')
    plt.tight_layout()
    plt.show()

    print("\n--- 2. Линейный график (Среднее количество очков по годам) --- ")
    avg_pts_by_year = df_champions.groupby('Year')['PTS'].mean().reset_index()
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=avg_pts_by_year, x='Year', y='PTS', marker='o')
    plt.title('Среднее количество очков по годам')
    plt.xlabel('Год')
    plt.ylabel('Среднее количество очков')
    plt.grid(True)
    plt.show()


    print("\n--- 3. Столбчатые диаграммы --- ")
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df_champions, x='Win')
    plt.title('Количество игр по результатам (0: Поражение, 1: Победа)')
    plt.xlabel('Результат игры')
    plt.ylabel('Количество игр')
    plt.xticks(ticks=[0, 1], labels=['Поражение', 'Победа'])
    plt.show()

    avg_pts_by_team = df_champions.groupby('Team')['PTS'].mean().sort_values(ascending=False).head(5).reset_index()
    plt.figure(figsize=(12, 6))
    sns.barplot(data=avg_pts_by_team, x='Team', y='PTS', palette='viridis')
    plt.title('Топ-5 команд по среднему количеству очков')
    plt.xlabel('Команда')
    plt.ylabel('Среднее количество очков')
    plt.show()

    print("\n--- 4. Boxplot (Распределение очков в зависимости от победы/поражения) --- ")
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df_champions, x='Win', y='PTS')
    plt.title('Распределение очков в зависимости от результата игры')
    plt.xlabel('Результат игры (0: Поражение, 1: Победа)')
    plt.ylabel('Очки')
    plt.xticks(ticks=[0, 1], labels=['Поражение', 'Победа'])
    plt.show()

    print("\n--- 5. Heatmap корреляций --- ")
    correlation_matrix = df_champions.corr(numeric_only=True)
    plt.figure(figsize=(16, 12))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
    plt.title('Матрица корреляций между числовыми признаками')
    plt.show()
else:
    print("DataFrame 'df_champions' не найден. Пожалуйста, загрузите данные сначала.")
```
